According to the previous heatmap, we could easily find that, targeting at the `popularity` variable, some variables have a comparatively higher correlation: `liveness`, `instrumentalness`, and `acousticness`.. Thus, we added them one by one to the model and check the significance.

In order to smooth our single variable regression model and save time, we first initiated a method for singluar variable regression. The functions take two columns of the dataset and generate a model based on them. Then, we will ask it to print the model's r^sqaure in order to compare among them. As mentioned at the beginning of this MusicIn1970s part, we will take the independent variable with the highest r^square to be our first variable to add in the regression model since it enables we explain more data points.


```python
def singleReg(x, y):
    x1 = music_1970s[x];
    y1 = music_1970s[y];
    my_linear_model = lr() 
    my_linear_model.fit(x1.values.reshape(-1, 1), y1) 
    x1new = x1.values.reshape(-1, 1)
    predictY = my_linear_model.predict(x1new)
    SSR = sum((y1 - predictY)**2)
    SST = sum((y1 - np.mean(y1))**2)
    r_square = 1 - SSR / SST
    print(" ")
    print(y + " ~ " + x + "｜")
    print("                               r^square: {}".format(np.round(r_square, 6)), end =" ")
```

```python
from sklearn.linear_model import LinearRegression as lr

singleReg('mode', 'popularity')
singleReg('explicit', 'popularity')
singleReg('acousticness', 'popularity')
singleReg('danceability', 'popularity')
singleReg('energy', 'popularity')
singleReg('duration_ms', 'popularity')
singleReg('instrumentalness', 'popularity')
singleReg('valence', 'popularity')
singleReg('tempo', 'popularity')
singleReg('liveness', 'popularity')
print("✔️", end =" ")
singleReg('loudness', 'popularity')
singleReg('speechiness', 'popularity')
```

Thus, for the first model we will select to join in our model is the `liveness`. Since in the `popularity` ~ `liveness` regression model, we have a r^square of 0.009882, which is comparatively largest among the other 11 models. Thus, it states that this model with `liveness` could better explain the `possibility` dependent variable since more data points in the data set are reached through this model, which is nearly 1%.


```python
my_linear_model = lr() 
x1 = music_1970s['liveness'];
y1 = music_1970s['popularity'];
my_linear_model = lr() 
my_linear_model.fit(x1.values.reshape(-1, 1), y1)
print('   Estimated slope: ', my_linear_model.coef_)
print('   Estimated intercept: ',my_linear_model.intercept_) 
```

Thus, as the first step completed, we got a model:

## Limitations

1. <b>[Spotify's music library]</b> Many songs are not available on Spotify due to the copyrights, courntries' regulations, sensitive contents, and many other factors. Thus, this dataset might not include all the songs that are publicated during 1970 to 2019. It might lead to some inaccuracies in results and our final result cannot be that applicable and useful for all past songs and all future songs. To possibly solve this problem, I believe combining my analysis to more datasets from other music streaming services like Apple Music, NetEase Music, MelOn music, and so on.

2. <b>[Algorithm for `popularity`]</b> Spotify changes each song's popularity based on the daily active listeners and monthly active liseners. Thus, if the creator updates the newest list of songs' popularity, our data research project's outcome might be changed since we are mainly examining the popularity and the possible factors related to it. It might lead to some miscalculation in results so that the final result might not be that accurate. The possible remedies here is to first generate a model/results based on the current dataset, then we can put the model over the up-to-dated datasets and check for validity.

3. <b>[Spotify's availability]</b> Some countries banned the usage of Spotify but the song's real audience and popularity might be larger. For example, China banned Spotify but there are many Chinese songs on Spotify, so the main audience for these songs are Chinese overseas, which is much smaller than the actual audience scale. It might lead to some underestimates in results and the final result might be applicable for these countries' music product. The possible remedy here is similar to the first problem, which is to have other analysis on the datasets provided by other streaming service.

4. <b>[Multiple-verison songs]</b> Many songs might have different verison, like feat. version, club version, and remastered version. But the dataset all counted them in one song. It might lead to some misleading and inaccurate results so our final result might not be useful for the multiple-version songs.

3. I make edits and cleaning on the `data.csv`.     - Original data shape: 19 Column * 174389 Rows
    - Cleaned data shape: 19 Column * 96757 Rows
         - pseudo code: select all `year` column of a value under 1970 or larger than 2019, drop the rows that contain the selected values
    

```python
## Code for data cleaning step#3
genre_list = data['year']

# delete observations before 1970 (1920-1969)
genreRom_list = np.where(genre_list < 1970)
for i in genreRom_list:
     data = data.drop(data.index[i])

# delete observations after 2019 (2020-2021)
genre_list = data['year']
genreRom_list = np.where(genre_list > 2019)
for i in genreRom_list:
     data = data.drop(data.index[i])
        
print(data.shape)

# Please uncomment the following row to see the first few columns (used for tracking)
# data.head()
```
